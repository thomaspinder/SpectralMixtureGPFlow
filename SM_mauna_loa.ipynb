```python
import numpy as np
import matplotlib.pyplot as plt
from kernel import SpectralMixture
from utils import sm_init
from gpflow.utilities import print_summary
from gpflow.kernels import Linear
from gpflow.models import GPR
from gpflow.optimizers import Scipy
from sklearn.datasets import fetch_openml
import tensorflow as tf

np.random.seed(123)
```

```python
def load_mauna_loa_atmospheric_co2():
    ml_data = fetch_openml(data_id=41187)
    months = []
    ppmv_sums = []
    counts = []

    y = ml_data.data[:, 0]
    m = ml_data.data[:, 1]
    month_float = y + (m - 1) / 12
    ppmvs = ml_data.target

    for month, ppmv in zip(month_float, ppmvs):
        if not months or month != months[-1]:
            months.append(month)
            ppmv_sums.append(ppmv)
            counts.append(1)
        else:
            # aggregate monthly sum to produce average
            ppmv_sums[-1] += ppmv
            counts[-1] += 1

    months = np.asarray(months).reshape(-1, 1)
    avg_ppmvs = np.asarray(ppmv_sums) / counts
    return months, avg_ppmvs


X, y = load_mauna_loa_atmospheric_co2()
y = y.reshape(-1, 1)
```

```python
Xtr, ytr = X[:300, :], y[:300, :]
Xte, yte = X[300:, :], y[300:, :]

fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(Xtr, ytr, label="Training points")
ax.plot(Xte, yte, label="Test points")
```

## Initialise spectral mixture kernel

```python
Q = 10
weights, means, scales = sm_init(train_x=Xtr, train_y=ytr, num_mixtures=Q)
k_sm = SpectralMixture(
    n_mixtures=Q,
    mixture_weights=weights,
    mixture_scales=scales,
    mixture_means=means + 1e-6,
)
```

## Define model

```python
kern = k_sm + Linear()
m = GPR((Xtr, ytr), kernel=kern)
print_summary(m, fmt="notebook")
```

## Optimise

```python
opt = tf.optimizers.Adam(learning_rate=0.005)
logfs = []

for i in range(5000):
    opt.minimize(m.training_loss, m.trainable_variables)
    loss = -m.training_loss().numpy()
    if i % 100 == 0:
        print(f"Step {i}: {loss}")
    logfs.append(loss)

plt.plot(logfs)
```

```python
print_summary(m, fmt="notebook")
```

## Predict

```python
mu, sigma = m.predict_y(X)
mun, sigman = mu.numpy().ravel(), sigma.numpy().ravel()

fig, ax = plt.subplots(figsize=(14, 7))
ax.plot(Xtr, ytr, "o", alpha=0.5, label="Training points")
ax.plot(Xte, yte, label="Truth", color="blue")
ax.plot(X, mun, label="Predictive mean", color="#b5121b")
ax.fill_between(
    X.ravel(),
    mun - sigman,
    mun + sigman,
    alpha=0.5,
    label="predictive variance",
    color="#b5121b",
)
ax.legend(loc="best")
```
